Hay varias operaciones que puedes realizar en las cadenas cuando las manipulamos. En este ejercicio, usarás métodos de cadena para modificar el texto con hechos sobre la Luna y luego extraerás información para crear un breve resumen.

El texto con el que trabajarás es el siguiente:

```python
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. 
The highest daylight temperature of the Moon is 127 C."""
```

Primero, divide el texto en cada oración para trabajar con su contenido:

```python
oracion=text.split('.')
print(oracion)
```

Ahora, define algunas palabras clave para búsqueda que te ayudarán a determinar si una oración contiene un hecho.

```python
# Define las palabras pista: average, temperature y distance suenan bien
keyWords=['average','temperature','distance']
result=[]
```

Cre un bucle para imprimir solo datos sobre la Luna que estén relacionados con las palabras clave definidas anteriormente:

```python
# ciclos para imprimir los datos relacionados
for ora in oracion:# recorre la lista que contiene las oraciones
  for key in keyWords:# recorre las palabras clave
    search=ora.find(key)# search sirve para guardar el resultado de buscar las plabras claves en las oraciones
    if search!=-1:# en caso de encontrar las keywords en la oracion
      print(ora)# visualizara la oracion en pantalla
```

Finalmente, actualiza el bucle(ciclo) para cambiar C a Celsius.

```python
# Ciclo para cambiar C a Celsius por medio de replace
for ora in text:
  res=text.replace('C','Celsius')
print(res)
```
